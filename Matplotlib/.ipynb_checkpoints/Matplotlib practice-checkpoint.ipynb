```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y)
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y)
plt.title('First Graph')
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x, y)
plt.title('First Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# Add int values on the axis
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x, y)

plt.title('First Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])
plt.show()
```

```python
# Add color and label
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x, y, color='red', label='2x')

plt.title('First Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
# Control line width
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x, y, color='red', label='2x', linewidth=5)

plt.title('First Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
# short hand notation
# fmt = '[color][marker][line]'
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x, y, 'b^--')

plt.title('First Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.show()
```

```python
# Line number two
x = [1,2,3,4]
y = [5,6,7,8]
x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2, x2**2, 'r')
plt.plot(x, y)

plt.title('First Graph')

plt.show()
```

```python
# Resize your graph
x = [1,2,3,4]
y = [5,6,7,8]
plt.figure(figsize=(2.5,2.5), dpi=200)

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2, x2**2, 'r')
plt.plot(x, y)

plt.title('First Graph')

plt.show()
```

```python
# Save your graph
x = [1,2,3,4]
y = [5,6,7,8]
plt.figure(figsize=(2.5,2.5), dpi=200)

x2 = np.arange(0, 4.5, 0.5)
plt.plot(x2, x2**2, 'r')
plt.plot(x, y)

plt.title('First Graph')

plt.savefig('mygraph.png')

plt.show()
```

# BarChart

```python
labels = ['A', 'B', 'C']
values = [1,2,3]

plt.bar(labels, values)

plt.show()
```

```python
# Add hatch
labels = ['A', 'B', 'C']
values = [1,2,3]

bars = plt.bar(labels, values)

bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.show()
```
