```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y)
```

```python
x = [1,2,3]
y = [2,4,6]
plt.plot(x, y)
plt.title('First Graph')
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x, y)
plt.title('First Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
# Add int values on the axis
x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x, y)

plt.title('First Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])
plt.show()
```
